```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
##hypterparamater tuning
```

```python

n_estimators=[x for x in np.arange(100, 1201, 100)]
max_depth=[x for x in np.arange(5,21,5)]
max_features=["auto","sqrt"]
min_samples_split=[2,5,10,20,50,100]
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
##creating random grid

random_grid={"n_estimators":n_estimators,
             "max_depth":max_depth,
             "max_features":max_features,
             "min_samples_split":min_samples_split,
             "min_samples_leaf":min_samples_leaf
    
    
}
```

```python
print(random_grid)
```

```python
rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring="neg_mean_squared_error", n_iter=10,
                            cv=5, verbose=2, random_state=234, n_jobs=1)
```

```python
rf_random.fit(xtrain, ytrain)
```

```python
predict=rf_random.predict(xtest)
```

```python
predict
```

```python
#3 check predictions
sns.displot(ytest-predict)
```

```python
sns.jointplot(ytest, predict)
```

```python
## store the model as pickle file
import pickle
file=open("random_forest_regression_model.pkl", "wb")
pickle.dump(rf_random, file)
```
